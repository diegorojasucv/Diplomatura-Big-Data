````python
import os
import tarfile
from six.moves import urllib

FETCH_URL = "http://dblandit.com/datasets/housing.tgz"
SAVE_PATH = os.path.join("datasets", "housing")

def fetch_and_extract_data(fetch_url=FETCH_URL, save_path=SAVE_PATH):
    if not os.path.isdir(save_path):
        print("Creating path '" + save_path + "'...")
        os.makedirs(save_path)
    tgz_path = os.path.join(save_path, "housing.tgz")
    print("Saving dataset from", fetch_url, "to '" + tgz_path + "'...")
    urllib.request.urlretrieve(fetch_url, tgz_path)
    print("Extracting data to '", save_path, "'...")
    housing_tgz = tarfile.open(tgz_path)
    housing_tgz.extractall(path=save_path)
    housing_tgz.close()
    print("Done.")
````

# Test Set #

Creemos un test set separando 20% del dataset de forma random:

````python
import numpy as np
def split_train_set(data_set, test_ratio):
    # Crear un array con todos los indices del dataSet y permutarlos
    shuffled_indices = np.random.permutation(len(data_set))
    
    # Separar los indices en test_Set y train set segun el test_ratio.
    test_set_size = int(len(data_set) * test_ratio)
    test_indices = shuffled_indices[:test_set_size]
    train_indices = shuffled_indices[test_set_size:]
    
    # Dividir el dataset segun indices de train
    train_set, test_set = data_set.iloc[train_indices], data_set.iloc[test_indices]
    print(len(train_set), "train +", len(test_set), "test")
    
    return train_set, test_set
````

Para usar la funcion:

````python
train_set, test_set = split_train_set(housing, 0.2)
````

A lo largo de varias corridas, se usará el dataset completo.

Ejemplo:

````python
import hashlib

def test_set_check(identifier, test_ratio, hash=hashlib.md5):
    return hash(np.int64(identifier)).digest()[-1] < 256 * test_ratio
````

````python
def split_train_test_by_id(data, test_ratio, id_column):
    ids = data[id_column]
    in_test_set = ids.apply(lambda id_: test_set_check(id_, test_ratio))
    return data.loc[~in_test_set], data.loc[in_test_set]
````

Como el dataSet no tiene una columna de Id creamos una:

````python
# agrego columna `index`
housing_with_id = housing.reset_index()

train_set, test_set = split_train_test_by_id(housing_with_id, 0.2, "index")
````

# Custom Transformers #

Respetar la API de Scikit para garantizar la integracion:
 - Crear una clase
 - Implementar metodos
   - `fit()` devuelve `self`
   - `transform()`
   - `fit_transform()` = fit + transform [+ optimizaciones]
   
Ejemplo:

````python
from sklearn.base import BaseEstimator, TransformerMixin
import numpy as np

rooms_ix, bedrooms_ix, population_ix, household_ix = 3, 4, 5, 6

class CombinedAttributesAdder(BaseEstimator, TransformerMixin):
    def __init__(self, add_bedrooms_per_room = True):
        self.add_bedrooms_per_room = add_bedrooms_per_room
    
    def fit(self, X, y=None):
        return self # nothing else to do
    
    def transform(self, X, y=None):
        rooms_per_household = X[:, rooms_ix] / X[:, household_ix]
        people_per_household = X[:, population_ix] / X[:, household_ix]
        if self.add_bedrooms_per_room:
            bedrooms_per_room = X[:, bedrooms_ix] / X[:, rooms_ix]
            return np.c_[X, rooms_per_household, people_per_household,bedrooms_per_room]
        else:
            return np.c_[X, rooms_per_household, people_per_household]

attr_adder = CombinedAttributesAdder(add_bedrooms_per_room=False)
housing_extra_attribs = attr_adder.transform(housing.values)
````

Hyperparametros: `add_bedrooms_per_room` con default `True`.

Permitir descubrir si este atributo ayuda o empeora el aprendizaje del algoritmo. 

Mediante los hyperparametros se puede agregar paso de preparacion de datos de los que no se este seguros. Cuando mas automatizado este este proceso mas fácil sera probar diversas combinaciones.

### Transformation Pipilines ###

Ejemplo de union de pipile:

````python
from sklearn.base import BaseEstimator, TransformerMixin

class DataFrameSelector(BaseEstimator, TransformerMixin):
    def __init__(self, attribute_names):
        self.attribute_names = attribute_names
    def fit(self, X, y=None):
        return self
    def transform(self, X):
        return X[self.attribute_names].values
````

````python
from sklearn.pipeline import FeatureUnion
from sklearn.pipeline import Pipeline
from sklearn.preprocessing import StandardScaler

num_attribs = list(housing_num)
cat_attribs = ["ocean_proximity"]

num_pipeline = Pipeline([
        ('selector', DataFrameSelector(num_attribs)),
        ('imputer', SimpleImputer(strategy="median")),
        ('attribs_adder', CombinedAttributesAdder()),
        ('std_scaler', StandardScaler()),
    ])

cat_pipeline = Pipeline([
        ('selector', DataFrameSelector(cat_attribs)),
        ('label_binarizer', LabelBinarizer()),
    ])

full_pipeline = FeatureUnion(transformer_list=[
        ("num_pipeline", num_pipeline),
        ("cat_pipeline", cat_pipeline),
    ])

housing_prepared = full_pipeline.fit_transform(housing)
housing_prepared
````

### RandomForestRegressor ###

Explicar algoritmo y mencionar `Ensemble Learning`.

````python
from sklearn.ensemble import RandomForestRegressor
from sklearn.model_selection import cross_val_score

modelo_forestReg = RandomForestRegressor()
modelo_forestReg.fit(housing_prepared, housing_labels)

housing_predictions = modelo_forestReg.predict(housing_prepared)

# Error training set
modelo_forestReg_mse = mean_squared_error(housing_labels, housing_predictions)
modelo_forestReg_rmse = np.sqrt(modelo_forestReg_mse)
print('modelo_forestReg_rmse:', modelo_forestReg_rmse, '\n')

# Error cross validaton
scores = cross_val_score(modelo_forestReg, housing_prepared, housing_labels, 
                         scoring="neg_mean_squared_error", cv=10)
print('x-validation scores:', scores)

forest_rmse_scores = np.sqrt(-scores)
display_scores(forest_rmse_scores)
````

- Error disminuyó!
