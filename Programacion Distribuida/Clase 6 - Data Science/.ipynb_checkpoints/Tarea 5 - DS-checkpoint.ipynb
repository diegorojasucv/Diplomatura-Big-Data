#### 40) De cada producto, ¿Cuál es el máximo, mínimo y promedio vendidos? Ordenar la lista por nombre de producto. 

```python
orders.groupby('item_name').agg({'quantity':['max', 'min', 'mean']}).sort_values(by = 'item_name')
```
